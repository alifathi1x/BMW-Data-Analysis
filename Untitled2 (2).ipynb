# Show datas on plot:

```python
sales = df.groupby("year")["price"].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=sales,x="year", y="price",markers="o")
plt.title("BMW sales per year")
plt.xlabel("Year")
plt.ylabel("price")
plt.grid()
plt.xticks(sales["year"], rotation=45)
plt.show()
```

```python
# نتیجه گیری : رشد قابل توجه شرکت از سال 1996
```

# Comparing the company's sales from 2010 to 2020

```python
data = {
    "year":[2010,2020],
    "amount":[82120,2168516]
}
df = pd.DataFrame(data)
print(df.describe())
plt.figure(figsize=(10, 6))
plt.title("BMW sales since 2010-2020")
plt.plot(df["year"], df["amount"],marker="o")
plt.xlabel("Year")
plt.ylabel("amount")
plt.grid()
plt.show()
```

# # Comparing bmw sales in IRAN & AMERICA

```python
x = [2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
y1 = [860000, 745999, 512333, 312000, 55442, 664780, 99000, 1236548, 342100, 409083, 109087, 475999, 123987]
y2 = [1122308, 1179317, 1302774, 1203482, 1043829, 1236989, 1440315, 1547057, 1699853, 1838268, 1933647, 2002998,
      2123947]

plt.plot(x, y1, color="red", label="IRAN")
plt.plot(x, y2, color="blue", label="USA")

plt.title("Comparison of BMW sales In Iran-USA")
plt.xlabel("Year")
plt.ylabel("amount")
plt.legend()
plt.grid()
plt.show()
```

```python
# result ----------------> bmw sales in AMERICA is higher than IRAN 
# نتیجه گیری : مقایسه فروش شرکت در دو کشور ایران و امریکا

# because sanctions and inflation growths:
#نموداز نزولی شرکت در ایران به دلیل تحریم ها و رشد تورم


years = list(range(2005, 2024))
inflation_rate = [1, 1.03, 1.06, 1.09, 1.12, 1.15, 1.18, 1.21, 1.24, 1.27,
                  1.30, 1.33, 1.36, 1.39, 1.42, 1.45, 1.48, 1.51, 1.54]
inflation_growth = [rate * 1.86 for rate in inflation_rate]
plt.figure(figsize=(10, 6))
plt.plot(years, inflation_growth, marker='o', linestyle='-', color='b')
plt.title("187 percent inflation growth since 2005")
plt.xlabel("Year")
plt.ylabel("Inflation growth")
plt.grid(True)
plt.show()
```

```python
# %187 inflation growth in Iran
# نتیجه گیری : رشد 187 درصدی تورم در ایران
```

# comparing diffrent ages for each bmw company products:

```python
np.random.seed(0)
ages_18_40 = np.random.randint(18, 41, 100)
ages_40_70 = np.random.randint(40, 71, 50)
data = pd.DataFrame({
    "Age": np.concatenate((ages_18_40, ages_40_70)),
    "Group": ["18-40"] * len(ages_18_40) + ["40-70"] * len(ages_40_70)
})
sns.pairplot(data, hue='Group', height=5)
plt.title("Comparison of the BMW sales for 18-40 years and 40-70")
plt.show()
```

```python
# beetween (30-45) has been the most sold out products
# نتیجه گیری: افزایش میزان فروش بین سنین 30 تا 45 سال بیشتر است
```

# Comparing Sales By Gender

```python
data = {
    'Gender': ["Male", 'Female'],
    'Purchases': [120, 150]
}

df = pd.DataFrame(data)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Purchases', data=df, palette='Set2')
plt.title('Comparison of Purchases by Gender for BMW Products')
plt.xlabel('Gender')
plt.ylabel('Number of Purchases')
plt.show()
```
